```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib as mp
```

Ci-dessus, il faut importer les différents modules afin de travailler les données. 
En dessous, je vais importer le fichier csv qui contient les données que je veux analyser. 
J'emploie 'AI' qui désigne 'Accès Information'

```python
ai = pd.read_csv('ati.csv')
```

```python
ai
```

J'ai mis 'AI' seul afin de voir si les données se sont importées correctement. 

```python
ai.year
```

La première cellule que je vais travailler est la cellule des années. Je vais donc utiliser value_counts afin de voir combien il y a de demandes dans les années inscrites dans le document.

```python
ai.year.value_counts()
```

Avec cette commande, on voit qu'il y a eu davantage de demandes d'accès en 2016 qu'en 2015. La commande suivante va transformer cette information en pourcentage. 

```python
ai.year.value_counts() / len(ai.year) * 100 
```

Donc, 52% des demandes d'accès à l'information ont eu lieu en 2016. On voit aussi qu'environ 4% des demandes d'accès à l'information depuis 2015 ont été réalisé depuis le début de cette année. 

Pour finir avec les années, je vais faire un petit graphique afin d'illustrer l'information que nous venons de découvrir. 

```python
ai.year.value_counts().plot(kind='pie',title='Graphique des demandes par années')
```

La deuxième cellule que je veux travailler est la cellule owner_org_title. 

```python
ai.owner_org_title.value_counts()
```

En effectuant un value_counts, on remarque que c'est la défense nationale qui a obtenu le plus de demandes d'accès à l'information. Je vais vouloir isoler les données qui concernent la défense nationale dans une liste

```python
defense = ai.owner_org_title == 'National Defence | Défense nationale'
ai[defense]
```

On lance le script pour être certain que nous avons seulement les données sont bien isolées. Pour conclure cette cellule, nous allons mettre les demandes d'accès de la défense nationale dans un diagramme à bande, car c'est plus jolie.

```python
ai[defense].year.value_counts().plot(kind='bar', title='Demandes à la défense nationale')
```

Comme pour les demandes en général, il y a eu plus de demandes en 2016 à la défense nationale. 

La troisième cellule que je veux travaille est le nombre de pages.

```python
ai.pages
```

D'abord, je veux savoir la médiane du nombres de pages afin de séparer les données en deux moitiés.

```python
ai.pages.median()
```

On apprend que la moitié des demandes à reçu moins de 11 pages (donc l'autre moitié plus de 11 pages). J'ai aussi fait une addition du nombre de pages totals, mais je n'utiliserai pas cette donnée présentement. 

```python
ai.pages.sum()
```

```python
ai.pages.value_counts()
```

Donc, avec la médiane j'ai déterminé qu'il y avait la moitié des documents qui avaient moins de 11 pages. J'ai fait un value_counts pour voir comment était distribué le nombre de pages par document. Afin de faire un nouveau tableau relativement simple à lire, je vais prendre les données pour les documents qui comportent 10 pages et moins. 

```python
peudepages = ai.pages < 11
ai[peudepages] 
```

Je vais maintenant combiner ma nouvelle liste des documents avec 10 pages et moins à ma liste de la défense nationale que j'ai créé précédement. 

```python
ai[peudepages & defense]
```

Pour terminer, je vais à nouveau faire un tableau pour illustrer la répartition des documents de 10 pages et moins qui concernent la défense nationale.  

```python
ai[peudepages & defense].pages.value_counts().plot(kind='bar', title='Nombre de pages par document de la défense')
```

On peut facilement voir qu'il y a eu une majorité de demandes d'accès qui n'ont pas porté fruit, car il n'y a eu aucune page d'attribuée à celles-ci.
